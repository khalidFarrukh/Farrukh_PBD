```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
path = "D:\project_PBD.csv"
df = pd.read_csv(path, header=None)
```

```python
headers = ["Country","City","Date","AvgTemperature"]
```

```python
df.columns = headers
```

```python

df.replace("?",np.nan, inplace=True)
```

```python
df["Country"].isnull().sum()
```

```python
df["Country"].replace(np.nan, 'Pakistan', inplace=True)
```

```python
df.head(15)
```

```python
df["City"].isnull().sum()
```

```python
df.shape
```

```python
df.dropna(subset=["City"], axis=0, inplace=True)
df.reset_index(drop=True, inplace=True)
```

```python
df.shape
```

```python
df.dtypes
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df.dtypes
```

```python
Islamabad = df[df["City"]=='Islamabad']
```

```python
Islamabad.set_index('Date', inplace=True)
```

```python
Islamabad.head(10)
```

```python
Karachi = df[df["City"]=='Karachi']
```

```python
Karachi.set_index('Date', inplace=True)
```

```python
Karachi.head(10)
```

```python
plt.ylabel('Avg Temperature per day', fontsize=20)
plt.xlabel('Year', fontsize=20)
plt.title('Islamabad yearly temperature from 1995 to 2020',fontsize=20)
Islamabad['AvgTemperature'].plot(figsize=(24,16), linewidth=0.4, fontsize=20)
plt.show()
```

```python
plt.ylabel('Avg Temperature per day', fontsize=20)
plt.xlabel('Year', fontsize=20)
plt.title('Karachi yearly temperature from 1995 to 2020',fontsize=20)
Karachi['AvgTemperature'].plot(figsize=(24,16), linewidth=0.4, fontsize=20,color='orange')
plt.show()
```

```python
df_rm = pd.concat([Islamabad['AvgTemperature'], Karachi['AvgTemperature']], axis=1)
df_rm.plot(figsize=(24,16), linewidth=0.4, fontsize=20)
plt.title('Daily Avg temperature of Karachi and Islamabad from 1995 to 2020',fontsize=20)
plt.xlabel('Year', fontsize=20);
```
